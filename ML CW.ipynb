```python
x1 = df.drop(['diagnosis', 'id'], axis=1)
x1.head()
```

```python
y1 = df['diagnosis']
y1.head()
```

```python
pd.value_counts(y1).plot.bar(color = "red")
plt.title('Before SMOTE')
plt.xlabel('Class')
plt.ylabel('Frequency')
y1.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE 

x_train, x_test, y_train, y_test = train_test_split(x1,y1,test_size=0.3, random_state = 2)
```

```python
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier


svm_clf=svm.SVC(kernel='linear',gamma = 'auto', C=2)
svm_clf.fit(x_train,y_train)
y_pred_svm = svm_clf.predict(x_test)
print("SVM Accuracy:",metrics.accuracy_score(y_test, y_pred_svm))

sgd_clf = SGDClassifier(loss="hinge", penalty="l2", max_iter=5)
sgd_clf.fit(x_train, y_train)
y_pred_sgd = sgd_clf.predict(x_test)
print("SGD Accuracy:",metrics.accuracy_score(y_test, y_pred_sgd))

dt_clf = DecisionTreeClassifier()
dt_clf.fit(x_train,y_train)
y_pred_dt = dt_clf.predict(x_test)
print("DT Accuracy:",metrics.accuracy_score(y_test, y_pred_dt))

mlp_clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
mlp_clf.fit(x_train, y_train)
y_pred_mlp = mlp_clf.predict(x_test)
print("MLP Accuracy:",metrics.accuracy_score(y_test, y_pred_mlp))

rf_clf = RandomForestClassifier(max_depth=2, random_state=0)
rf_clf.fit(x_train, y_train)
y_pred_rf = rf_clf.predict(x_test)
print("RF Accuracy:",metrics.accuracy_score(y_test, y_pred_rf))

```

```python
print('SVM')
print(classification_report(y_test,y_pred_svm))
print('SGD')
print(classification_report(y_test,y_pred_sgd))
print('DT')
print(classification_report(y_test,y_pred_dt))
print('MLP')
print(classification_report(y_test,y_pred_mlp))
print('RF')
print(classification_report(y_test,y_pred_rf))
```

```python
sm = SMOTE(random_state=42)
x_res, y_res = sm.fit_resample(x_train,y_train)
pd.value_counts(y_res).plot.bar(color = 'g')
plt.title('SMOT')
plt.xlabel('Class')
plt.ylabel('Frequency')
```

```python
svc = SVC(kernel="linear")

pipeline_svm = Pipeline([('Feature_selection',RFECV(estimator=svc,step=1,scoring="accuracy")), 
                        ('scaler1',MinMaxScaler()),
                        ('pca1', PCA(n_components=2)),
                        ('SVM',svm.SVC(kernel='sigmoid',gamma = 'auto', C=2))])

pipeline_MLP =Pipeline([('scaler2',MinMaxScaler()),
                        ('Feature_selection',RFECV(estimator=svc,step=1,scoring="accuracy")),
                        ('pca2', PCA(n_components=2)),
                        ("MLP", MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1))])

pipeline_SGD =Pipeline([('Feature_selection',RFECV(estimator=svc,step=1,scoring="accuracy")),
                        ('scalar3',MinMaxScaler()),
                        ('pca3', PCA(n_components=2)),
                        ("SGD",SGDClassifier(loss="hinge", penalty="l2", max_iter=10))]) 

pipeline_RF = Pipeline([('Feature_selection',RFECV(estimator=svc,step=1,scoring="accuracy")),
                        ('scalar4',MinMaxScaler()),
                        ('pca4', PCA(n_components=2)),
                        ("RF",RandomForestClassifier(max_depth=4, random_state=2))])

pipeline_DT = Pipeline([('Feature_selection',RFECV(estimator=svc,step=1,scoring="accuracy")),
                        ('scalar5',MinMaxScaler()),
                        ('pca5', PCA(n_components=2)),
                        ("DT",DecisionTreeClassifier(max_depth=4))])
```

```python
pipelines = [pipeline_svm, pipeline_MLP, pipeline_SGD, pipeline_RF, pipeline_DT]
pipe_dict ={0: 'SVM', 1: 'MLP', 2: 'SGD', 3:'RF', 4:'DT'}

for pipe in pipelines:
    pipe.fit(x_res, y_res)

for i,model in enumerate(pipelines):
    print("{} Test Accuracy: {}". format(pipe_dict[i],model.score(x_test,y_test)))
    metrics.accuracy_score
```
